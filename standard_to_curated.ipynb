```python
import pandas as pd
```

```python
standard_path = "standard/"
```

```python
standard_streamings = pd.read_parquet(standard_path + "streaming_history_standard")
```

```python
standard_streamings
```

```python
streaming_history = standard_streamings.rename(columns = {
    "endTime":"END_TIME",
    "artistName":"ARTIST_NAME",
    "trackName":"TRACK_NAME",
    "msPlayed":"MS_PLAYED"}
    )

streaming_history['END_TIME'] = pd.to_datetime(streaming_history['END_TIME'])

streaming_history['END_TRACK_DATE'] = streaming_history['END_TIME'].dt.date
streaming_history['END_TRACK_DATE'] = streaming_history['END_TRACK_DATE'].astype("datetime64")

streaming_history['END_TRACK_TIME'] = streaming_history['END_TIME'].dt.strftime('%H:%M')

streaming_history = streaming_history.drop(['END_TIME'], axis=1)
```

```python
streaming_history
```

```python
def save_to_csv(df, csv_name):
    str_name = "curated/" + csv_name
    df.to_csv(str_name, sep = ";")
```

```python
save_to_csv(streaming_history, "streaming_history_curated")
```
